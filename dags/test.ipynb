```python
import requests

# replace the "demo" apikey below with your own key from https://www.alphavantage.co/support/#api-key
url = f'https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol={ticker}&interval=60min&apikey={API_KEY}'
r = requests.get(url)
data = r.json()

print(data)
```

```python
type(data['Time Series (60min)'])
```

```python
len(data['Time Series (60min)'])
```

```python
data['Time Series (60min)']
```

```python
import pandas as pd
df = pd.DataFrame.from_dict(data['Time Series (60min)'], orient='index', dtype=float)
df
```

```python
df.reset_index(inplace=True)
df
```

```python
#change the column names
df.columns = ['timestamp','open', 'high', 'low', 'close', 'volume']
df
```

```python
df['close'] = df['close'].astype(float)
df['daily_returns'] = df['close'].pct_change()
df
```
